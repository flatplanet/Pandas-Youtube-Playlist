```python
import pandas as pd
import numpy as np
from numpy.random import randn
%matplotlib inline

# Create Dataframe
my_df = pd.DataFrame(randn(25,4), columns=["Mon", "Tues", "Wed", "Thurs"])
my_df
```

```python
# bar plot
my_df.abs().plot(kind='bar', stacked=False)
```

```python
# bar plot Stacked
my_df.abs().plot(kind='bar', stacked=True)
```

```python
# Shading 
my_df.abs().plot(kind='bar', stacked=True, alpha=0.4)
```

```python
# Title
my_df.abs().plot(kind='bar', stacked=True, alpha=0.4, title="My Awesome Bar Plot")
```

```python
# legend
my_df.abs().plot(kind='bar', stacked=True, alpha=0.4, title="My Awesome Bar Plot", legend=False)
```

```python
# Second Way
my_df.abs().plot.bar()
```

```python
# Second Way Stacked
my_df.abs().plot.bar(stacked=True)
```

```python
# Second Way Shading 
my_df.abs().plot.bar(stacked=True, alpha=0.4)
```

```python
# Second Way Title
my_df.abs().plot.bar(stacked=True, alpha=0.4, title="My Awesome Bar Graph")
```

```python
# Second Way Legend
my_df.abs().plot.bar(stacked=True, alpha=0.4, title="My Awesome Bar Graph", legend=False)
```

```python
# Grid Lines
my_df.abs().plot.bar(stacked=True, alpha=0.4, title="My Awesome Bar Graph", legend=False, grid=True)
```
