```python
import pandas as pd
import numpy as np
# import csv file
my_df = pd.read_csv('dog_data_short.csv')
my_df
```

```python
# Add Column From List
gender = ["Male","Female","Male","Male", "Female"]
my_df['Gender'] = gender
my_df
```

```python
# Add Default Values
my_df['Alive'] = [True]*len(my_df)
my_df
```

```python
# Add Nan Value
my_df["ShowDog"] = [np.nan]*len(my_df)
my_df
```

```python
# add column with .insert() - Allows Position
my_df.insert(1, "Adopted", [True]*len(my_df), True)
my_df
```

```python
# Add column with .assign() - creates new df
my_df2 = my_df.assign(Horse=[False]*len(my_df))
my_df2
```
