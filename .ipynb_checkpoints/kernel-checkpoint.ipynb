```python
import pandas as pd
import numpy as np
from numpy.random import randn
%matplotlib inline

# Create Dataframe
my_df = pd.DataFrame(randn(500,4), columns=["Mon", "Tues", "Wed", "Thurs"])
my_df
```

```python
# Density and kernel density estimation plots with scipy
my_df.plot(kind='kde')
```

```python
# KDE second way to learn shift tab
my_df.plot.kde(alpha=0.4, lw=4)
```

```python
# Density Plot
my_df.plot(kind='density') 
```

```python
# Other method
my_df.plot.density()
```

```python
# Title
my_df.plot.density(title="My Awesome Density Plot") 
```

```python
# Grid
my_df.plot.density(grid=True) 
```
