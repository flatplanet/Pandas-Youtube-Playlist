```python
import pandas as pd
import numpy as np
from numpy.random import randn
%matplotlib inline

# Create Dataframe
my_df = pd.DataFrame(randn(25,4), columns=["Mon", "Tues", "Wed", "Thurs"])
my_df
```

```python
# Line plot
my_df.plot(kind='line')
```

```python
# line plot absolute
my_df.abs().plot(kind='line')
```

```python
# Line single
my_df['Mon'].abs().plot(kind='line')
```

```python
# Line More than one
my_df[['Mon', 'Tues']].abs().plot(kind='line')
```

```python
# Size
my_df[['Mon', 'Tues']].abs().plot(kind='line', figsize=(10,5))
```

```python
# line Width
my_df[['Mon', 'Tues']].abs().plot(kind='line', figsize=(10,5), lw=3)
```

```python
# Title
my_df[['Mon', 'Tues']].abs().plot(kind='line', figsize=(10,5), title="My Awesome Line Graph")
```

```python
# Shading 
my_df[['Mon', 'Tues']].abs().plot(kind='line', figsize=(10,5), title="My Awesome Line Graph", alpha=0.3)
```

```python
# legend
my_df[['Mon', 'Tues']].abs().plot(kind='line', figsize=(10,5), title="My Awesome Line Graph", legend=False)
```

```python
# Second Way chart
my_df.plot.line()
```
