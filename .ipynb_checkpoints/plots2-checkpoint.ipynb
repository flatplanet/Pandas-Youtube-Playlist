```python
import pandas as pd
import numpy as np
from numpy.random import randn
%matplotlib inline

# Create Dataframe
my_df = pd.DataFrame(randn(500,4), columns=["Mon", "Tues", "Wed", "Thurs"])
my_df
```

```python
# Boxplots
my_df.plot(kind='box')
```

```python
# Boxplots Size
my_df.plot(kind='box', figsize=(10,5))
```

```python
# Boxplots Color
my_df.plot(kind='box', figsize=(10,5), cmap="jet")
```

```python
# No Legend
my_df.plot(kind='box', figsize=(10,5), legend=True)
```

```python
# Title
my_df.plot(kind='box', figsize=(10,5), title='My Awesome Box Plot', cmap='winter')
```

```python
# Other Way
my_df.plot.box(figsize=(10,5), title='My Awesome Box Plot', cmap='winter')
```

```python
# Add Grid
my_df.plot.box(figsize=(10,5), title='My Awesome Box Plot', cmap='winter', grid=True)
```
