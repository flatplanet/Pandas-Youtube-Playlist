```python
import pandas as pd
import numpy as np
from numpy.random import randn
%matplotlib inline

# Create Dataframe
my_df = pd.DataFrame(randn(500,4), columns=["Mon", "Tues", "Wed", "Thurs"])
my_df
```

```python
# Hexplots
my_df.plot(kind='hexbin', x='Mon', y='Tues')
```

```python
# Hexplots gridsize
my_df.plot(kind='hexbin', x='Mon', y='Tues', gridsize=10) # next try 25
```

```python
# Hexplots color
my_df.plot(kind='hexbin', x='Mon', y='Tues', gridsize=25, cmap='magma') 
```

```python
# Other method
my_df.plot.hexbin(x='Mon', y='Tues', gridsize=25) 
```

```python
# Title
my_df.plot.hexbin(x='Mon', y='Tues', gridsize=25, title="My Awesome Hexbin Plot") 
```

```python
my_df.plot.hexbin(x='Mon', y='Tues', gridsize=25, grid=True) 
```
