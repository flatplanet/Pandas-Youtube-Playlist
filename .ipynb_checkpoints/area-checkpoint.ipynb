```python
import pandas as pd
import numpy as np
from numpy.random import randn
%matplotlib inline

# Create Dataframe
my_df = pd.DataFrame(randn(25,4), columns=["Mon", "Tues", "Wed", "Thurs"])
my_df
```

```python
# area plot
my_df.abs().plot(kind='area', stacked=False)
```

```python
# Shading
my_df.abs().plot(kind='area', alpha=0.4)
```

```python
# Title
my_df.abs().plot(kind='area', alpha=0.4, title="My Awesome Area Plot", table=True)
```

```python
# legend
my_df.abs().plot(kind='area', alpha=0.4, title="My Awesome Area Plot", table=True, legend=False)
```
