```python
import pandas as pd
import numpy as np
# Create Data
stuff = {
    'Corporation':['Apple', 'Google', 'Meta', 'Apple', 'Google','Meta'],
    'Employees':['John', 'April', 'Wes', 'Beth', 'Justin', 'Steph'],
    'Salary':[200, 220, 190, 130, 120, 150],
}
# Create Dataframe
my_df = pd.DataFrame(stuff)
my_df
```

```python
# Create Function
def times10(x):
    #return x * 100
    return format (x * 1000, ',d')
```

```python
# Use Apply Function
my_df["Salary"].apply(times10)
```

```python
# Append to current DF
my_df["Salary"] = my_df["Salary"].apply(times10)
my_df
```

```python
# Function on multiple columns
def namer(x):
    return 'Codemy: ' + x
            
```

```python
# Apply to one Column
pd.DataFrame(my_df["Employees"].apply(namer))
```

```python
# Apply To Two Columns
my_df[["Corporation", "Employees"]].apply(namer)
```

```python
# Append to current Dataframe
my_df[["Corporation", "Employees"]] = my_df[["Corporation", "Employees"]].apply(namer)
my_df
```

```python
def namer2(x):
    for i in x:
        if i == "Apple":
            return "Codemy!"
        else:
            return x
        
```

```python
my_df[["Corporation", "Employees"]] = my_df[["Corporation", "Employees"]].apply(namer2)
my_df
```
