```python
import pandas as pd
import numpy as np
from numpy.random import randn
%matplotlib inline

# Create Dataframe
my_df = pd.DataFrame(randn(500,4), columns=["Mon", "Tues", "Wed", "Thurs"])
my_df
```

```python
# Scatter plot
my_df.plot(kind='scatter', x='Mon', y='Tues', c="Wed", cmap='magma') # matplotlib colormaps
```

```python
# https://matplotlib.org/stable/tutorials/colors/colormaps.html
my_df.plot(kind='scatter', x='Mon', y='Tues', c="Wed", cmap='PuBu')
```

```python
# By Size
my_df.plot(kind='scatter', x='Mon', y='Tues', s=my_df['Wed'])
```

```python
# By Size easier to read
my_df.plot(kind='scatter', x='Mon', y='Tues', s=my_df['Wed']*50)
```

```python
# By Size easier to read with Alpha
my_df.plot(kind='scatter', x='Mon', y='Tues', s=my_df['Wed']*100, alpha=0.4)
```

```python
# line Width
my_df[['Mon', 'Tues']].abs().plot(kind='line', figsize=(10,5), lw=3)
```

```python
# Title
my_df[['Mon', 'Tues']].abs().plot(kind='line', figsize=(10,5), title="My Awesome Line Graph")
```

```python
# Shading 
my_df[['Mon', 'Tues']].abs().plot(kind='line', figsize=(10,5), title="My Awesome Line Graph", alpha=0.3)
```

```python
# legend
my_df[['Mon', 'Tues']].abs().plot(kind='line', figsize=(10,5), title="My Awesome Line Graph", legend=False)
```

```python
# Second Way chart
my_df.plot.line()
```
