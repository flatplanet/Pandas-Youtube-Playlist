```python
import pandas as pd
import numpy as np
from numpy.random import randn
%matplotlib inline

# Create Dataframe
my_df = pd.DataFrame(randn(500,4), columns=["Mon", "Tues", "Wed", "Thurs"])
my_df
```

```python
# Hexplots
my_df.plot(kind="hexbin", x="Mon", y="Tues")
```

```python
# Hexplots gridsize
my_df.plot(kind="hexbin", x="Mon", y="Tues", gridsize=10)
```

```python
# Hexplots color
my_df.plot(kind="hexbin", x="Mon", y="Tues", gridsize=10, cmap="Wistia")
```

```python
# Other method
my_df.plot.hexbin(x="Mon", y="Tues", C="Wed", gridsize=10)
```

```python
# Title
my_df.plot.hexbin(x="Mon", y="Tues", C="Wed", gridsize=10, title="My Awesome Hex Plot")
```

```python
# Grid
my_df.plot.hexbin(x="Mon", y="Tues", C="Wed", gridsize=10, title="My Awesome Hex Plot", grid=True)
```
