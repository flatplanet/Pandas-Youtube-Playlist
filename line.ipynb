```python
import pandas as pd
import numpy as np
from numpy.random import randn
%matplotlib inline

# Create Dataframe
my_df = pd.DataFrame(randn(25,4), columns=["Mon", "Tues", "Wed", "Thurs"])
my_df
```

```python
# Line plot
my_df.plot(kind="line")
```

```python
# line plot absolute
my_df.abs().plot(kind="line")
```

```python
# line Width
my_df.abs().plot(kind="line", lw=5)
```

```python
# Line single
my_df['Mon'].abs().plot(kind="line")
```

```python
# Line More than one
my_df[['Mon', 'Tues']].abs().plot(kind="line")
```

```python
# Size
my_df[['Mon', 'Tues']].abs().plot(kind="line", figsize=(10,5))
```

```python
# Title
my_df[['Mon', 'Tues']].abs().plot(kind="line", title="My Awesome Line Chart")
```

```python
# Shading 
my_df[['Mon', 'Tues']].abs().plot(kind="line", alpha=0.3)
```

```python
# legend
my_df[['Mon', 'Tues']].abs().plot(kind="line", legend=False)
```

```python
# Second Way chart
my_df.plot.line()
```
